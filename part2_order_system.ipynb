```python
# Menu data
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# Print by category
categories = ["Starters", "Mains", "Deserts"]

for cat in categories:
  print("===", cat, "===")

  for item in menu:
    if menu[item]["category"] == cat:
      if menu[item]["available"]: status = "Available"
      else:
        status = "Unavailable"

      print(item, "₹", menu[item]["price"], "[", status, "]")

# Total items
print("Total items:", len(menu))

# Available items
count = 0
for item in menu:
  if menu[item]["available"]:
    count += 1
print("Available items:", count)

# Most expensive
max_item = max(menu, key=lambda x: menu[x]["price"])
print("Most expensive:", max_item, menu[max_item]["price"])

# Items under 150
print("Under 150:")
for item in menu:
  if menu[item]["price"] < 150:
    print(item, menu[item]["price"])

```

```python
# Create empty cart
cart = []
```

```python
# Function to add item
def add_item(name, qty):

  # Check item exists
  if name not in menu:
    print("Item not found")
    return

  # Check availability
  if not menu[name]["available"]:
    print("Item not available")
    return

  # If already in cart - Update qty
  for item in cart:
    if item["item"] == name:
      item["quantity"] += qty
      print("update quantity")
      return

  # Add new item
  cart.append({"item":name,
               "quantity": qty,
               "price": menu[name]["price"]
               })
  print("Item added")
```

```python
# Function to remove item
def remove_item(name):

  for item in cart:
    if item["item"] == name:
      cart.remove(item)
      print("Item removed")
      return
  print("Item not in cart")
```

```python
# Function to update quantity
def update_qty(name, qty):

  for item in cart:
    if item["item"] == name:
       item["quantity"] = qty
       print("Quantity update")
       return

  print("Item not found")
```

```python
# Simulate actions

add_item("Paneer Tikka", 2)
print(cart)

add_item("Gulab Jamun", 1)
print(cart)

add_item("Paneer Tikka", 1)  # Should be updated
print(cart)

add_item("Mystery Burger", 1) # Do not exist

add_item("Chicken Wings", 1)  # Unavailable

remove_item("Gulab Jamun")
print(cart)
```

```python
# Print final bill
print("==== Order Summary ====")

subtotal = 0

for item in cart:
    total = item["quantity"] * item["price"]
    subtotal += total

print(item["item"], "x",item["quantity"],"₹", total)

print("-------------------")

print("Subtotal:", subtotal)

# Calculate gst
gst = subtotal * 0.05
print("GST:", round(gst, 2))

# Final amount
total_pay = subtotal + gst
print("Total:", round(total_pay, 2))

```

```python
# Copy inventory
inv_copy = {}

for item in inventory:
  inv_copy[item] = inventory[item].copy()
```

```python
# Reduce stock using cart
for itemn in cart:

  # If item is string - convert to correct format
  if isinstance(item, str):
    name = item
    qty = 1                 # Assume 1
  else:
      name = item["item"]
      qty = itemn["quantity"]

if name in inv_copy:
    inv_copy[name]["stock"] -= qty
```

```python
# Check reorder items
print("Low Stock Items:")

for item in inv_copy:
  if inv_copy[item]["stock"] <= inv_copy[item]["reorder_level"]:
    print(item, "->", inv_copy[item]["stock"])
```
